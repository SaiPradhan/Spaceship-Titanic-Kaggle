```python
def split_cabin_deck(st):
    tokens=st.split('/')
    return tokens[0]

def split_cabin_num(st):
    tokens=st.split('/')
    return int(tokens[1])

def split_cabin_side(st):
    tokens=st.split('/')
    return tokens[2]

print(split_cabin_deck('B/0/P'))
```

```python
train2['Deck']=train2['Cabin'].apply(split_cabin_deck)
train2
```

```python
train2['Num']=train2['Cabin'].apply(split_cabin_num)
train2
```

```python
train2['side']=train2['Cabin'].apply(split_cabin_side)
train2
```

```python
g4=train2.groupby('Deck')
for d, data in g4:
    print(d)
    print(data.Transported.value_counts())
```
